What factors affect the COVID-19's Impact on Airport Traffic?

 Data Summary: numeric variables: PercentOfBaseline, Geography
          categorical variable: AirportName

```python
import pandas as pd
import numpy as np
```

```python
Airport_impact = pd.read_csv('https://raw.githubusercontent.com/YihangSun123/AIM-5001/master/AIM-5001/covid_impact_on_airport_traffic.csv')
Airport_impact
```

```python
Airport_impact['City'].describe()
```

```python
city_plot = Airport_impact['City'].value_counts().plot.bar(title = 'Frequency Distribution of city')
city_plot
```

```python
data = Airport_impact[['PercentOfBaseline', 'City', 'Geography']]
data
```

```python
data.plot.scatter(x = 'PercentOfBaseline', y = 'City')
```

conclusion：The level of covid-19 impact to airplort is related city. Many things can affect the covid-19 impact to airplort in different city, such as population and leve of covid-19.

```python
data1 = Airport_impact[['Date', 'PercentOfBaseline', 'AirportName']]
data1
```

```python
data1['AirportName'].describe()
```
