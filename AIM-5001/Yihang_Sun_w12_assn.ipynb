# XML

```python
import urllib.request
from lxml import objectify
```

```python
path, headers = urllib.request.urlretrieve('https://raw.githubusercontent.com/YihangSun123/AIM-5001/master/AIM-5001/Untitled-1.xml')
parsed = objectify.parse(open(path))
root = parsed.getroot()
```

```python
data = []
for elt in root.book:
    el_data = {}
    for child in elt.getchildren():
        el_data[child.tag] = child.pyval
    data.append(el_data)
```

```python
perf = pd.DataFrame(data)
perf.head()
```
